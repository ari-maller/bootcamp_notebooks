```python
ds = yt.load_sample('medium_tipsy')
```

```python
sp = ds.sphere('max',(10,'kpc'))
```

```python
yt.ProjectionPlot(ds,'x',('gas','density'), 
                  center='max',width=(50,'kpc'))
```
