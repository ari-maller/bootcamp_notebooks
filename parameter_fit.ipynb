```python
import numpy as np
rng = np.random.default_rng(seed=42)
#random uniform from 0 to 1
rng.random(5)
```

```python
rng.integers(low=0,high=10,size=5)
```

```python
rng.normal(loc=0.0,scale=1.0,size=5)
```

```python
x=rng.normal(size=50000)
import matplotlib.pyplot as plt
_=plt.hist(x,range=[-2,2],bins=20)
```

```python
def linear(x,m=1.0,b=0.0):
    return m*x+b

#line with noise
x=rng.random(size=50)
y=linear(x)+rng.normal(scale=0.2,size=x.size)
plt.scatter(x,y)
```

```python
p,cov=np.polyfit(x,y,1,w=1/0.2+0.*x,cov=True)
```

```python
xrange=np.linspace(0,1)
plt.scatter(x,y)
plt.plot(xrange,p[0]*xrange+p[1])
```
